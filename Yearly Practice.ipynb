```python
# 1. Ride Frequency
ride_counts = df21['member_casual'].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(ride_counts, labels=ride_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Member Type-2021 ")
plt.show()
print(ride_counts)
```

```python
# 1. Ride Frequency
ride_counts = df22['member_casual'].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(ride_counts, labels=ride_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Member Type-2022")
plt.show()
print(ride_counts)
```
