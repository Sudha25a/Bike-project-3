```python
# Calculate total rides
total_rides = len(df21)

# Calculate casual rides
casual_rides = len(df21[df21['member_casual'] == 'casual'])

# Calculate casual rider percentage
casual_rider_percentage = (casual_rides / total_rides) * 100

print(f"Casual Rider Percentage: {casual_rider_percentage:.2f}%")

```

```python
# Group by month and calculate casual rider percentage for each month
monthly_casual_ridership = df21.groupby(['year', 'Month'])['member_casual'].apply(lambda x: (x == 'casual').sum() / len(x) * 100).reset_index(name='casual_percentage')
```

```python
# Plot monthly casual rider percentage
Month_order= ['January', 'February','March', 'April', 'May', 'June', 'July', 'August', 'september','October', 'November', 'December' ]

plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='casual_percentage', hue='year', data=monthly_casual_ridership, marker='o', order= 'Month_order')
plt.title('Monthly Casual Rider Percentage')
plt.xlabel('Month')
plt.ylabel('Casual Rider Percentage')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.tight_layout()
plt.show()
```

```python
# Group by month and calculate casual rider percentage for each month
monthly_casual_ridership22 = df22.groupby(['year', 'Month'])['member_casual'].apply(lambda x: (x == 'casual').sum() / len(x) * 100).reset_index(name='casual_percentage')
```

```python
# Plot monthly casual rider percentage
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='casual_percentage', hue='year', data=monthly_casual_ridership22, marker='o', color= 'b')
plt.title('Monthly Casual Rider Percentage')
plt.xlabel('Month')
plt.ylabel('Casual Rider Percentage')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.tight_layout()
plt.show()
```

```python
# Group by month and calculate casual rider percentage for each month
monthly_casual_ridership23 = df23.groupby(['year', 'Month'])['member_casual'].apply(lambda x: (x == 'casual').sum() / len(x) * 100).reset_index(name='casual_percentage')
```

```python
# Plot monthly casual rider percentage
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='casual_percentage', hue='year', data=monthly_casual_ridership23, marker='o', color= 'b')
plt.title('Monthly Casual Rider Percentage')
plt.xlabel('Month')
plt.ylabel('Casual Rider Percentage')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.tight_layout()
plt.show()
```
